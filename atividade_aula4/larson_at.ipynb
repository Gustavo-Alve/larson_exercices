## Importando minhas bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lista_numerica = [128, 100, 180, 150, 200, 90, 340, 105, 85, 270,
    200, 65, 230, 150, 150, 120, 130, 80, 230, 200,
    110, 126, 170, 132, 140, 112, 90, 340, 170, 190]

df = pd.DataFrame(lista_numerica, columns=['Dados'])

minimo = min(df['Dados']) #65
maximo = max(df['Dados']) #340
amplitude = (maximo - minimo) #275
mediana= df['Dados'].median() #145.0
moda = df['Dados'].mode() # 150 - 200
media = df['Dados'].mean() # 159.44
classes = pd.cut(lista_numerica, bins=([65, 105, 145, 185, 225, 265, 305, 345])) #esse numero de bins serve para separar minhas classes de onde começam e onde terminam
frequencia = classes.value_counts()
total_frequencia  = frequencia.sum() # 30 porem no meu codigo esta dando 29 pois ele pega a primeira classe como 65-105 e assim por diante
intervalos = classes.categories
ponto_medio = (intervalos[0].left + intervalos[0].right - 1)/2 # apos Calcular o primeiro ponto medio  e apenas somar mais intervalo de classes que e 40 nesse caso
frequencia_relativa = frequencia/30 # A frequencia relativa e minha frequencia dividida pelo total dela que no exercicio e 30, faça isso para todas as frequencias
# print(frequencia_relativa.round(2)) #round usado para pegar apenas dois valores apos a virgula
frequencia_acumulada = frequencia.cumsum() #cumsum() funçao usada para calcular frequencia acumulada. os valores estao um pouco divergentes devido aquele problema das classes
# print(ponto_medio)

```
